```python
import os
import numpy as np
import pickle
import matplotlib.pyplot as plt

def load():
    data_contents = []

    with open("fft_test.pickle", 'rb') as f:
        while True:
            try:
                data_contents.append(pickle.load(f))
            except EOFError:
                break
    return np.array(data_contents)

def plot(name, measurement, rng=5000):
    fft = np.fft.fft(measurement, axis=0)
    fig, axs = plt.subplots(1, 2, figsize=(20, 5))
    axs[0].set_title("FFT")
    axs[0].plot(fft)
    axs[0].set_ylim([-rng, rng])
    axs[0].set_xlim([40, 500])

    axs[1].set_title("Measurement")
    axs[1].plot(measurement)

    fig.suptitle(name, fontsize=16)
    fig.show()

data = load()
```

```python
sensor_1_ambient = data[0]['data']
plot('Ambient Philips Hue', sensor_1_ambient)
```

```python
sensor_1_led = data[1]['data']
plot('Cheap LED', sensor_1_led, 20000)
```

```python
sensor_1_strobe = data[2]['data']
plot('Strobing LED', sensor_1_strobe, 40000)
```
